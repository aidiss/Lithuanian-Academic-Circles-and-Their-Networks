#Įvadas

Šiandieninė mokslo sistema yra labai sudėtingas reiškinys, kurį galima nagrinėti įvairiais pjuviais. Vienas iš pjūvių - (tarpdiscipliniškumas). Disertacijos gynimas kuriuo pažymimas doktoranto tapimas mokslininku. Ši situacija pagrindžia susidomėjimą discipliniškumo raiška disertacijos gynimo procese.

Lietuvos aukštojo mokslo sistemoje mokslinę veiklą įprasta klasifikuoti pasitelkiant trijų lygių sistemą. Aukščiausioji sistemos dalis - sritys, kurios skirstomos į kryptis, o šios į šakas. Mokslo kryptys paprastai koreliuoja su ilgametėmis disciplininomis, tokiomis kaip teisė, teologija, filosofija ir medicina. Būtent šis lygis mums atrodo tinkamas mokslo struktūros nagrinėjimui.

Yra nemažai bandymų parodyti mokslo struktūrą apskritai. Įvairius bandymus apžvelgia ir mokslo struktūrų žemėlapių panašumus ir skirtumus aptaria Klavans ir Boyack (2009). Naudodami jų tyrimų rezultatus sulyginsime disertacijų gynimo krypčių žemėlapį su didesnio pasauline mokslo struktūra. 

Konceptualų instrumentarijų tikslesniam tarpdiscipliniškumo formos įvardinimui siūlo Klein (2010). Jos tyrimai apibendrina įvairias tarpdiscipliniškumo formas. Labiausiai ji išskiria daugiadiscipliniškumą (multidisciplinary), tarpdiscipliniškumą (interdisciplinarity) ir trandiscipliniškumą (transdisciplinarity). Tyrime bandysime įžvelgti, kiek įmanoma atpažinti tarpdiscipliniškumo darinius disertacijos gynimų praktikoje.

Mikulskienė su Mažrimiene (2013) aptaria savireguliaciją doktorantūros gynimo procese. Straipsnyje aptariamos doktorantūros komiteto teisės ir pareigos. Atlikta tarpinstituciškumo bei tarpdiscipliniškumo analizė. Skirtingai nuo Mikulskienos ir Mažrimienės darbo, šiame straipsnyje visas dėmesys yra skiriamas tarpdiscipliniškumui. Daugiau dėmesio skiriama analizei pasitelkiant grafas bei vizualizacijai.

Leydesdorff (2007, 2011, 2013) ir Silva (2013) naudodami socialinių tinklų analizės metodus bandė įvertinti mokslinių žurnalų tarpdiscipliniškumą. Remsimės jų tyrimo rezultatais, tačiau iki tol reikės palyginti mokslinį žurnalą ir discipliną.

Apžvelgus literatūrą, galime daryti prielaidą, jog mokslo struktūros tyrimų pasitelkiant disertacijų gynimo nėra. Tuo tarpu mes galime šią spragą užildyti pateikdami Lietuvos mokslo struktūros žemėlapio dalį.

Porter (2009) nustatė, kad mokslas iš tikrųjų tampa vis labiau tarpdisciplinišku, tačiau tik mažais žingsniais. Straipsniai yra cituojami šalia esančių disciplinų.
Tarpdiscipliniškumo vertinimas didelia dalimi priklauso nuo to, iš kokios disciplinos į jį žvelgsime. Krishnan (2009) pristato įvairias perspektyvas: filosofinę, antropologinę, sociologinę, istorinę, vadybos. Jis taip pat pristato išlikimo strategijas. Klein (2010) išskiria daugiadiscipliniškumą (multidisciplinary), tarpdiscipliniškumą (interdisciplinarity) ir trandiscipliniškumą (transdisciplinarity). Siaurąja prasme tarpdiscipliniškumas yra tai, kas integruoja, sukuria kokybiškai naujus santykius tarp dviejų disciplinų. Daugiadiscipliniškumas, kai disciplinos yra šalia, tačiau jų turinys, metodai ir kitos savybės nėra sumaišomos, tarp jų nėra sukuriami realūs ryšiai.

Didžiausia šio tyrimo naujovė, tai, kad mokslo žemėlapis bus kuriamas naudojant doktorantūros gynimų dalyvių mokslo kryptis. Ieškant atitikmenų su labiau pieštais mokslo žemėlapiais mūsų pasirinkimą galime prigretinti prie mokslinės publikacijas. Mokslinis straipsnis yra artimas disertacijai. Disertacija paprastai yra autorinis, didesnės apimties darbas, tuo tarpu dauguma straipsnių yra mažesnių mokslinių tyrimų, rezultatas. Kitas skirtumas, mokslinis straipsnis yra mokslinio žurnalo, kuris talpina daug kitų publikacijų. Disertacija nėra talpinama rinkiniuose, šiuo atžvilgiu ji panašesnė į knygą, kuri skirta tik vieno autoriaus vienam darbui. Tačiau, kiekviena disertacija yra priskiriama tam tikrai mokslo krypčiai. Kita, panašiai kaip mokslinis straipsnis, disertacija turi praeiti peer-review procesą. Mokslinių straipsnių šis procesas yra labai įvairialypis, double-blind ir kiti metodai. Tuo tarpu disertacijos gynimas, bent jau Lietuvoje, yra pompastiškesnis ir skaidresnis procesas. Disertacijos be doktoranto dalyvauja dar bent 8 nariai (vadovas, 5 tarybos nariai ir oponentas). Esame įsitikine, jog recenzavimo proceso žemėlapiai yra tokie pat informatyvus kaip ir anksčiau aptartieji, turint duomenis juos būtų galima tikrai panaudoti. Tačiau, mokslinių publikacijų atveju procesas yra įslaptintas, tad nubrėžti mokslo žemėlapį naudojant mokslo straipsnių recenzavimo tinklus yra įmanoma, tačiau tam trūksta informacijos. Nepaisant to, kad disertacijos vadovo indėlis į disertacijos tyrinėjamos temos nagrinėjamą gali svyruoti, jį priskiriame kaip antro svarbumo objektu. #4. Reikalavimai mokslo krypčiai doktorantūrą reguliuojančioje teisėje.

Pirmojo laikotarpio doktorantūros nuostatuose teigiama, kad doktoranto vadovas - "mokslininkas, disertacijos tematika atliekantis mokslinius tyrimus", tačiau nerašoma, ar gynimo taryba turi būti disertacijos tematika atliekantis moksliniu tyrimus  (LRV, 2001)

"50. Oficialieji oponentai turi būti skirtingų institucijų mokslininkai, bent vienas – ginamos disertacijos mokslo krypties (šakos) ir bent vienas iš kitos (neturinčios bendros doktorantūros teisės) institucijos." (VU, 2003; KTU, 2003; VDU, 2003; LSMU, 2003; VGTU, 2003 ) 
2003 VU mokslo doktorantūros reglamente: [nuoroda](http://www.vu.lt/site_files/DRS/VU_DOKTORANTUROS_REGLAMENTAS__.pdf).

2010 m. LRS patvirtintuose  mokslo doktorantūros nuostatuose mokslinės krypties atitikimo reikalavimas doktoranto vadovui nepakito. Pabrėžtina, jog doktoranto vadovo tyrimai, jų kryptys, turi atitikti disertacijos tematiką. Tad aptikę neatitinkančias kryptis galime teigti, jog vadovas nuo savo paskutinio mokslinio laipsnio, jį paskyrusios doktorantūros komisijos, yra pajudėjęs kitos krypties link. Antra, pabržiama, jog gynimo tarybos apibrėžime nėra minima, jog turi atitikti kryptis. Tačiau, steigiant doktorantūros studijų doktorantūros komitetą, turi būti vertinamas "..komiteto narių atitinkamos mokslo krypties (šakos) ar krypčių (šakų) mokslinių tyrimų lygis"(6 str.) Tad gaunama situacija, kurioje doktorantūros komitetas turi atitikti mokslo krypties lygį, tačiau tolimesnis jų pasirinkimas nėra reglamentuojamas.

Buvo išnagrinėti trys universitetų mokslo doktoranturos reglamentai. Visi jie 2010 metų "8.2. Doktorantūros komitetas – aukšto lygio mokslininkų grupė, atsakinga už atitinkamos krypties (-čių) doktorantūros studijų bei mokslinių tyrimų vykdymą" (VU, 2010) daugiau paminėjimo nėra.

Išnagrinėjus reglamentus rasta, jog reikalavimų ir pageidavimų krypčiai yra palyginti nedaug. Pasitaikė šie reikalavimai: 1) stodamas turi pateikti vieju tos srities mokslininkų rekomendacijas; 2) vadovas turi būti tos mokslo srities arba toje srityje aktyviai veikiantis mokslininkas (KTU DR 2010,29); 3) gali būti skiriami kitos krypties konsultantai(KTU DR 2010,30); 4) vienas oponentas turi būtinai būti tos pačios krypties o kitas būtinai iš kitos; 4) gynimo taryba sudaroma iš tos krypties doktorantūros (KTU DR 2010,73)

2010 m. KTU(KTU, 2010) reglamente kryptys minimos šitais aspektais. Mokslo krypties doktorantūros komitetas sudaromas ir tvirtinamas senato, bet tik po to kai suderina  su LMT. Tarpinstitucinės doktorantūros atveju sudaroma kartu su kita institucija. Skiriasi formuluotė: "Vadovas turi būti doktorantūros mokslo krypties (šakos) mokslininkas." Pastebima, jog mokslinia konsultantai gali būti kitos krypties, bet nebūtinai. Turbūt svarbiausias reikalavimas, kurio nepasitaikė prieš tai apžvelgtame VU reglamente, kad "Tarybos nariai turi būti tos mokslo krypties (šakos) mokslininkai", tačiau jie negali turėti bendrų publikacijų. Toliau kalbama apie apie "mokslo krypčių sandūrą", pasak reglamento, du tarybos nariai *gali* būti kitų mokslo krypčių mokslininkai, bet tik tais atveju, "jeigu darbas atliktas mokslo krypčių sandūroje arba jeigu to reikia, kad būtų galima visapusiškai įvertinti ginamą darbą". 

Dažnu atveju komisijoje negalėjo dalyvauti bendraautoriai. Tai atmeta galimybę, kad komisijoje pasirodys doktoranto bendraautorius, su kuriuo doktorantas turėjo tarpdisciplinį ryšį. Tačiau, mūsų manymu tai iš esmės nekeičia mūsų sumanymo, tačiau ateičiai galima būtų žemėlapį patikslinti sukuriant ryšius tarp doktorantų ir kitose mokslo kryptyse tyrinėjimus atliekančių mokslininkų, su kuriais doktorantas turėjo bendrų publikacijų.

Galima teigti, jog klasifikacijos sistema atspindi Lietuvos mokslo tyrimų supratimą apie mokslo struktūrą. Per reikalavimus atsižvelgti į ŠMM reikalavimus ir per konsultacijas su ŠMM, per doktorantūros komisijos sudėtį, doktoranto studijų ir disertacijos priskyrimas tam tikrai krypčiai bei gynimo tarybos narių pakvietimas joje dalyvauti yra susijęs su mokslo krypčių veikla. T.y. vertindami doktorantūros gynimo dalyvių sudėtį galime įžvelgti Lietuvos mokslo struktūrą.

```python
# a simple function to compute hierarchical cluster on both rows and columns, and plot heatmaps
def heatmap(dm):
    from scipy.cluster.hierarchy import linkage, dendrogram
    from scipy.spatial.distance import pdist, squareform
    
    D1 = squareform(pdist(dm, metric='euclidean'))
    D2 = squareform(pdist(dm.T, metric='euclidean'))
    
    f = plt.figure(figsize=(8, 8))

    # add first dendrogram
    ax1 = f.add_axes([0.09, 0.1, 0.2, 0.6])
    Y = linkage(D1, method='complete')  #complete, single, average, weighted, centroid, median, ward
    Z1 = dendrogram(Y, orientation='right')
    ax1.set_xticks([])
    ax1.set_yticks([])

    # add second dendrogram
    ax2 = f.add_axes([0.3, 0.71, 0.6, 0.2])
    Y = linkage(D2, method='complete')
    Z2 = dendrogram(Y)
    ax2.set_xticks([])
    ax2.set_yticks([])

    # add matrix plot
    axmatrix = f.add_axes([0.3, 0.1, 0.6, 0.6])
    idx1 = Z1['leaves']
    idx2 = Z2['leaves']
    D = D1[idx1, :]
    D = D[:, idx2]
    im = axmatrix.matshow(D, aspect='auto', origin='lower', cmap='hot')
    axmatrix.set_xticks([])
    axmatrix.set_yticks([])
    
    return {'ordered' : D, 'rorder' : Z1['leaves'], 'corder' : Z2['leaves']}
heatmap(scatdf)
```
