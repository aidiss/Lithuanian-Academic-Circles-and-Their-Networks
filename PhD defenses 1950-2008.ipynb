```python
from functions import *
import pandas as pd
result = html_to_one_txt()
transition = [from_raw_to_clean(x) for x in result]
COLUMNS = ['dkt_vrd', 'tema', 'dkt_lps', 'data_lks', 'dkt_dsc', 'rng_ins']
df = pd.DataFrame([x[0] for x in transition], columns=COLUMNS )
df = tvarkyk(df) # 
df.head()
```

```python
def rsktem(tema):  #rask temą
    return df[df.tema.str.contains(tema, regex=True, case=False)]
def rskvrd(vardas):
    return df[df.dkt_vrd.str.contains(vardas, regex=True, case=False)]
```

```python
%pylab inline
import mpld3
import matplotlib.pyplot as plt
mpld3.enable_notebook()
#mpld3.disable_notebook()
pd.options.display.mpl_style = 'default'

plt.figure(1, figsize=(15,7))
for x in ['H','S','P','T','B']:
    plt.subplot(111)
    fltrd = df[df['dkt_dsc'].str.contains(x, na=False)]
    grpd = fltrd.groupby(fltrd['data_lks'].map(lambda x: x.year)).size()
    plt.plot(grpd.index, grpd.values,label='{}'.format(x),linewidth=4,alpha=0.7)
    
#plot(df.groupby(df['data_lks'].map(lambda x: x.year)).size())

plt.xlabel(u'Metai')
plt.ylabel(u'Gynimų sk')
plt.title(u'Disertacijų gynimai pagal metus')

legend(loc=2) #bbox_to_anchor=(0., 1.02, 1., .102) , ncol=1 , mode="expand"

#plt.axis([1940, 2014, 0, 500])
plt.grid(True)
plt.show()
```

```python
#mpld3.disable_notebook()
def draw1(sritis):
    plt.figure(1, figsize=(20,7))
    fltrd = df[df['dkt_dsc'].str.contains(sritis, na=False)]
    for kryptis in ['01', '02', '03', '04', '05', '06', '07', '08', '09',  '10']:
        fltrd1 = df[df['dkt_dsc'].str.contains(kryptis, na=False)]
        grpd = fltrd1.groupby(fltrd['data_lks'].map(lambda x: x.year)).size()
        plt.plot(grpd.index, grpd.values,label='{}{}'.format(sritis,kryptis),linewidth=4,alpha=0.7)
    
    plt.xlabel(u'Metai')
    plt.ylabel(u'Gynimų sk')
    plt.title(u'{} srities disertacijų gynimai pagal metus'.format(sritis))
    legend(loc=2)
    plt.show()
[draw1(x) for x in ['H','S','P','T','B']]
```
