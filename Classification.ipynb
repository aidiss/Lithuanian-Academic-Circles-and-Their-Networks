# NetworkX

```python
import networkx as nx
G = nx.Graph()
G.add_nodes_from(unq)
G.add_nodes_from(unq2)

tada = [[x, list(df[df[u'saka'] == x].kryptis.values)] for x in unq]
for x in tada:
    for z in x[1]:
        G.add_edge(x[0],z)
        
def gauktipa(x):
    if len(df[df[u'kryptis'] == x]) != 0:
        return u'kryptis'
    elif len(df[df[u'saka'] == x]) != 0:
        return u'saka'
    else:
        return u"nzn"

def gauksriti(x):
    if list(df[df[u'kryptis'] == x].sritis.unique())[0]:
        return list(df[df[u'kryptis'] == x].sritis.unique())[0]
    if list(df[df[u'sritis'] == x].sritis.unique())[0]:
        return list(df[df[u'sritis'] == x].sritis.unique())[0]
```

```python
for x in G.nodes():
    G.node[x][u'sritis'] = ''
    G.node[x][u'tipas'] = ''
    
    TIPAS = gauktipa(x)
    G.node[x][u'tipas'] = TIPAS
    if TIPAS == u'kryptis':
        SRITIS = gauksriti(x)
        G.node[x][u'sritis'] = SRITIS
```
